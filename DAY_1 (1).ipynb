# **Types of Operators**
*Arithmetic,Comparison(Relational),Assignment,Logical,Bitwise,Identity,Membership are the names of operators.*


```python
# 1. Arithmetic operators -- (+,-.*,/,//,%,**)
a = 4
b = 2
print('Addition :',a+b)
print('Subtraction :',a-b)
print('Multiplication :',a*b)
print('Division :',a/b)
print('Floor Division :',a//b)
print('Modulus :',a%b)    # Remender
print('Exponent :',a**b)  # a^b
```

```python
# arithematic operators ->
# + - * / // % **

# +
a = 10
b = 20
print(a + b) # 10 + 20

# -
c = 20
d = 10
print(c - d) # 20 - 10

# *
e = 10
f = 5
print(e * f) # 10 * 5

# /
num_1 = 20
num_2 = 2
print(num_1 /  num_2) # float

# //
num_3 = 20
num_4 = 2
print(num_3 // num_4) # float division -> integer

# %
num_5 = 10
num_6 = 3
print(num_5 % num_6)

# **
num_7 = 5
num_8 = 3
print(num_7 ** num_8)

# example (even or odd)
num9 = int(input("Enter the number: "))
num10 = 2
print("Even or Odd in form of 0 and 1 :")
print(num9 % num10) # 0 -> it's even || 1 -> it's odd
```

```python
# Relational opr -> True or False
# == , < , > , >= , <= , !=

# ==
num1 = 10
num2 = 11
print(num1 == num2)

# < -> less than
num1 = 10
num2 = 11
print(num1 < num2)

# > -> greater than
num1 = 20
num2 = 10
print(num1 > num2)

# >= -> greater than and equal to
num1 = 20
num2 = 10
print(num1 >= num2)

# <= -> less than and equal to
num1 = 20
num2 = 10
print(num1 <= num2)


# != -> not equal
num1 = 20
num2 = 10
print(num1 != num2)
```

```python
# Assignment opr
# = , += , -= , *= , /= , //= , %=

# = -> assing opr used to declare a variable

a = 10
b = a + 20
print(b)

# +=
a = 10
a += 20
print(a)

#Example of all the assignment operators
print('Example of all the assignment operator answers')
A = 58
A = A + 8
print('A1 : ', A)
A = A - 8
print('A2 : ',A)
A += 8
print('A3 : ',A)
A -= 8
print('A4 : ',A)
A *= 8
print('A5 : ',A)
A /= 8
print('A6 : ',A)
A %= 8
print('A7 : ',A)
A //= 8
print('A8 : ',A)
# A **= 8
# print('A9 : ',A)
```

```python
# Logical operators
# and , or , not

# and
a = 10
b = 20
print(a and b)


#Example
a = 10
b = -10
print(a or b)
b = 0
print(a or b ,'\n')

# Example for string concept
name1 = 'hello'
name2 = ''
print('hello or blank :',name1 or not name2)
print('not hello or blank :', not name1 or name2)
```
